```python
# Biblio
# https://www.geeksforgeeks.org/python-pandas-dataframe-resample/
# https://stackabuse.com/pandas-library-for-data-visualization-in-python/

# import the necessary libraries
import numpy as np
import pandas as pd
import os
from matplotlib import pyplot
```

```python
# Show current path directory
os.getcwd()
```

```python
# Read house1_frag.csv data
house2_data = pd.read_csv('data\House_2.csv', delimiter = ',', parse_dates=[0], index_col=0)
house2_data = house1_data.dropna()
house2_data_filtered = house2_data[['Appliance4']]
monthly_resampled_Appliance_ALL_H2 = house2_data_filtered.resample('1H').mean()
monthly_resampled_Appliance4_H2 = house2_data_filtered.Appliance4.resample('1H').mean() # 1H Hour, 1D Day, 6H Hour 4H
```

```python
# Show Appliance1 by Hour
monthly_resampled_Appliance4_H2 = monthly_resampled_Appliance4_H2[(monthly_resampled_Appliance4_H2.index < '2015-06-10') | (monthly_resampled_Appliance4_H2.index > '2015-06-17')]
monthly_resampled_Appliance4_H2[(monthly_resampled_Appliance4_H2.index < '2014-01-24') | (monthly_resampled_Appliance4_H2.index > '2014-03-06')]
```

```python
# House_1_16.csv ( NO DATA Line 273100) 
monthly_resampled_Appliance_graph_H2 = monthly_resampled_Appliance_ALL_H2[(monthly_resampled_Appliance_ALL_H2.index < '2015-06-10') | (monthly_resampled_Appliance_ALL_H2.index > '2015-06-17')]

# House_1_3.csv ( NO DATA Line 134624) 
monthly_resampled_Appliance_graph_H2 = monthly_resampled_Appliance_graph_H2[(monthly_resampled_Appliance_graph_H2.index < '2014-01-24') | (monthly_resampled_Appliance_graph_H2.index > '2014-03-06')]

# Show Filtered data
to_plot_H2 = monthly_resampled_Appliance_graph_H2.groupby(monthly_resampled_Appliance_graph_H2.index.hour).mean()

to_plot_H2
```

```python
# Show Graph
# Quit Time
interpolated_H2 = to_plot.interpolate(method='spline', order=2)
print(interpolated_H2.head(24))
interpolated_H2.plot()
pyplot.show()
```
