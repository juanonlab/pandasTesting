```python
# Biblio
# https://www.geeksforgeeks.org/python-pandas-dataframe-resample/
# https://stackabuse.com/pandas-library-for-data-visualization-in-python/

# import the necessary libraries
import numpy as np
import pandas as pd
import os
from matplotlib import pyplot
```

```python
# Show current path directory
os.getcwd()
```

```python
# Read house1_frag.csv data
house3_data = pd.read_csv('data\House_3.csv', delimiter = ',', parse_dates=[0], index_col=0)
house3_data = house3_data.dropna()
house3_data_filtered = house3_data[['Appliance7']]
monthly_resampled_Appliance_ALL_H3 = house3_data_filtered.resample('1H').mean()
monthly_resampled_Appliance7_H3 = house3_data_filtered.Appliance7.resample('1H').mean() # 1H Hour, 1D Day, 6H Hour 4H
```

```python
# Show Appliance1 by Hour
monthly_resampled_Appliance7_H3 = monthly_resampled_Appliance7_H3[(monthly_resampled_Appliance7_H3.index < '2015-06-10') | (monthly_resampled_Appliance7_H3.index > '2015-06-17')]
monthly_resampled_Appliance7_H3[(monthly_resampled_Appliance7_H3.index < '2014-01-24') | (monthly_resampled_Appliance7_H3.index > '2014-03-06')]
```

```python
# House_1_16.csv ( NO DATA Line 273100) 
monthly_resampled_Appliance_graph_H3 = monthly_resampled_Appliance_ALL_H3[(monthly_resampled_Appliance_ALL_H3.index < '2015-06-10') | (monthly_resampled_Appliance_ALL_H3.index > '2015-06-17')]

# House_1_3.csv ( NO DATA Line 134624) 
monthly_resampled_Appliance_graph_H3 = monthly_resampled_Appliance_graph_H3[(monthly_resampled_Appliance_graph_H3.index < '2014-01-24') | (monthly_resampled_Appliance_graph_H3.index > '2014-03-06')]

# Show Filtered data
to_plot_H3 = monthly_resampled_Appliance_graph_H3.groupby(monthly_resampled_Appliance_graph_H3.index.hour).mean()

to_plot_H3
```

```python
# Show Graph
# Quit Time
interpolated_H3 = to_plot_H3.interpolate(method='spline', order=2)
print(interpolated_H3.head(24))
interpolated_H3.plot()
pyplot.show()
```
