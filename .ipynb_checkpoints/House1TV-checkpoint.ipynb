```python
# Biblio
# https://www.geeksforgeeks.org/python-pandas-dataframe-resample/
# https://stackabuse.com/pandas-library-for-data-visualization-in-python/
# https://chrisalbon.com/python/data_wrangling/group_pandas_data_by_hour_of_the_day/

# import the necessary libraries
import numpy as np
import pandas as pd
import os
from matplotlib import pyplot
```

```python
# Show current path directory
os.getcwd()
```

```python
# Read house1_frag.csv data
house1_data = pd.read_csv('data\House_1.csv', delimiter = ',', parse_dates=[0], index_col=0)
house1_data = house1_data.dropna()
house1_data_filtered = house1_data[['Appliance8']]
monthly_resampled_Appliance_ALL = house1_data_filtered.resample('1H').mean()
monthly_resampled_Appliance8 = house1_data_filtered.Appliance8.resample('1H').mean() # 1H Hour, 1D Day, 6H Hour 4H
```

```python
# Show Appliance1 by Hour
monthly_resampled_Appliance8 = monthly_resampled_Appliance8[(monthly_resampled_Appliance8.index < '2015-06-10') | (monthly_resampled_Appliance8.index > '2015-06-17')]
monthly_resampled_Appliance8[(monthly_resampled_Appliance8.index < '2014-01-24') | (monthly_resampled_Appliance8.index > '2014-03-06')]
```

```python

# House_1_16.csv ( NO DATA Line 273100) 
monthly_resampled_Appliance_graph = monthly_resampled_Appliance_ALL[(monthly_resampled_Appliance_ALL.index < '2015-06-10') | (monthly_resampled_Appliance_ALL.index > '2015-06-17')]

# House_1_3.csv ( NO DATA Line 134624) 
monthly_resampled_Appliance_graph = monthly_resampled_Appliance_graph[(monthly_resampled_Appliance_graph.index < '2014-01-24') | (monthly_resampled_Appliance_graph.index > '2014-03-06')]

# Show Filtered data
to_plot = monthly_resampled_Appliance_graph.groupby(monthly_resampled_Appliance_graph.index.hour).mean()

to_plot
```

```python
# Show Graph
# Quit Time
interpolated = to_plot.interpolate(method='spline', order=2)
print(interpolated.head(24))
interpolated.plot()
pyplot.show()
```
